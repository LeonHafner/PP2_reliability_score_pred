```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mlp_scores = [0.009, 0.028, 0.033, 0.039, 0.093]
linreg_scores = [-0.025, 0.010, 0.044, 0.030, 0.069]

tasks = ["AAV low vs. high", "AAV two vs. many", "GB1 low vs. high", "GB1 two vs. rest", "Meltome"]
```

```python
x = np.arange(len(tasks))
width = 0.35
```

```python
plt.figure(figsize=(7, 6))
plt.bar(x - width/2, mlp_scores, width, label="MLP", color="lightblue", edgecolor="black")
plt.bar(x + width/2, linreg_scores, width, label="LinReg", color="lightcoral", edgecolor="black")
plt.axhline(y=0, color="black", linewidth=0.8, linestyle="--")

plt.ylabel("Spearman Correlation", fontsize=15)
plt.xticks(x, tasks, rotation=90, fontsize=12)
plt.legend(fontsize=12)
plt.ylim(-1, 1)
plt.tight_layout()
```
